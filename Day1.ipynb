```python
import pandas as pd
df = pd.read_excel("Practice datasets/sales data.xlsx")

```

```python
from sklearn.impute import SimpleImputer

target_column = 'device_type'

imputer = SimpleImputer(strategy='most_frequent')
df[target_column] = imputer.fit_transform(df[[target_column]]).ravel()
```

```python
df.isnull().sum()
```

```python
#getting data types

data_types = df.dtypes

print("Data Types:")
print(data_types)
```

```python
mixed_data = df['cost']

#create an empty list for non_numeric values
non_numeric_values = []

#checking variables' data types(cost)

for value in mixed_data:
    if isinstance(value, str) and not value.isnumeric():
        non_numeric_values.append(value)

print ("Non_numeric Values:")
print(non_numeric_values)
```

```python
#removing value that is not the of variable's datatype
#specify variable(column)
numeric_column ='cost'

#define value to search and remove
value_to_remove = 'XXX'

#create a boolean mask to identify rows where the specified value is in the column
mask =(df[numeric_column]== value_to_remove)

df = df[~mask]

print(df)
```

```python
#assigning datatype for 'cost'

df['cost'] = df['cost'].astype(float)

print(df.dtypes)
```

```python
#assigning datatype for 'date'

df['date'] = pd.to_datetime(df['date'])

print(df.dtypes)
```

```python
df.head()
```

```python
#dealing with duplicated values
#identify duplicate rows
duplicates = df[df.duplicated()]

print("Duplicates Rows:")
print(duplicates)

#remove duplicates
df_no_duplicates = df.drop_duplicates()

print("No duplicates:")
print(df_no_duplicates)
```

```python
cleaned_data = df_no_duplicates
```
